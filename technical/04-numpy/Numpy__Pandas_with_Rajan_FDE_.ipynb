# Performance Test
----

Objective: to add 1 million numbers using python vs numpy
--- one taking less time will be the winner.

```python
import numpy as np
import time

SIZE = 1000000

A1 = range(SIZE)
A2 = range(SIZE)

L1 = np.arange(SIZE)
L2 = np.arange(SIZE)

# python test

start = time.time()
result = [x+y for x,y in zip(A1,A2)]
stop = time.time()

print('Python 🐍 Took '  ,(stop-start) * 1000, ' ms ')

# Numpy Test

start = time.time()
result = L1 + L2
stop = time.time()

print('💻 Numpy took ' , (stop-start) * 1000 , ' ms ')

```

# Solving Linear Equations using Numpy
---
<pre>
Day1
----
2 Stocks (Google) and 3 Stocks ( Apple ) were Purchase for $110

Day2

4 Stocks (Google) and 1 Stock (Apple)   purchased for $130

Task: Find the Price for each Stock of Google & Appl.


```python
# 2g + 3a = 110
# 4g + 1a = 130
```

# Lets Solve using Numpy
---

```python
# step 1 - prepare the arrays  representing coefficients & Constancts

coeff =  np.array ( [ [ 2, 3] , [ 4 , 1 ] ]  )

const = np.array ( [ 110 , 130 ])
```

np.linalg.solve()

```python
result = np.linalg.solve( coeff , const )

google = result[0]
apple  = result[1]
```

```python
print(f'Google : {google} , Apple {apple}')
```
